# Normalization

```python
# http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html
from sklearn.preprocessing import MinMaxScaler
```

```python
m = MinMaxScaler(feature_range=(0,1))
```

```python
ms = m.fit(df[['Marketing Spend']])
ms
```

```python
df.head()
```

```python
df['Marketing Spend'] = ms.transform(df[['Marketing Spend']])
```

```python
df.head()
```

```python
df['Administration'] = m.fit_transform(df[['Administration']])
```

```python
df.head()
```

```python
df['Transport'] = m.fit_transform(df[['Transport']])
```

```python
df.head()
```

# Max Absolute Scaler

```python
df3 = pd.read_csv('supershops.csv')
df3.describe()
```

```python
sns.distplot(df3['Marketing Spend']) #before abs
```

```python
#value/maxvalue
```

```python
df3['Marketing Spend']=df3['Marketing Spend']/df3['Marketing Spend'].abs().max()
```

```python
sns.distplot(df3['Marketing Spend']) #after abs
```

```python
#Administration
```

```python
sns.distplot(df3['Administration'])
```

```python
df3['Administration']=df3['Administration']/df3['Administration'].abs().max()
```

```python
sns.distplot(df3['Administration'])
```
